```python
#mapping the values of 1 and 2 to yes and no respectively and
#printing the first 5 records from the dataset.

haberdata['status'] = haberdata['status'].map({1:'yes', 2:'no'})
haberdata.head() 
```

Bivariate Analysis

```python
haberdata.plot(kind = "scatter" , x = 'age' , y = 'nodes')
plt.title('Scatter Plot of age vs nodes')
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(haberdata, hue="status", height=5)\
.map(plt.scatter, "age", "nodes") \
.add_legend()
plt.title('Scatter Plot Using Seaborn of age vs nodes')
plt.show()
```

```python
#3D Pair Plot
sns.set_style("whitegrid");
sns.pairplot(haberdata,hue ='status',height = 3)
plt.show()
```

```python
# Box-Plot
sns.boxplot(x='status',y='age', data=haberdata)
plt.title("Box showing age vs status")
plt.show()
```

```python
# Box-Plot
sns.boxplot(x='status',y='nodes', data=haberdata)
plt.legend
plt.title("Box showing age vs nodes")
plt.show()
```

```python
# A violin plot 
sns.violinplot(x='status',y='age', data=haberdata, size=8)
plt.show()
```

```python
# A violin plot 
sns.violinplot(x='age',y='nodes', data=haberdata, height=8)
plt.show()
```

```python
#sns.set_style("whitegrid");
sns.FacetGrid(haberdata, hue="status", height=5)\
.map(sns.distplot, "age") \
.add_legend()
plt.title('Histogram of age')
plt.show()
```

```python
sns.FacetGrid(haberdata,hue='status',height = 5)\
.map(sns.distplot,'year')\
.add_legend();
plt.title('Histogram of year')
plt.show()
```

```python
# A violin plot 
sns.violinplot(x='status',y='nodes', data=haberdata, size=8)
plt.legend
plt.title("Violin Plot")
plt.show()
```
