# Seaborn Statistical Charts

# Displot

```python
sns.distplot(car_crashes.alcohol, rug = True, fit =
stats.gausshyper);
```

```python
sns.distplot(car_crashes.speeding, rug = True, fit =
stats.gausshyper);
```

```python
sns.distplot(car_crashes.ins_losses, rug = True, fit =
stats.gausshyper);
```

# Jointplot(scatterplot)

```python
# Scatterplot - Bivariate Distribution
sns.jointplot(x ="speeding", y= "alcohol", data= car_crashes);
```

```python
# Scatterplot - Bivariate Distribution
sns.jointplot(x ="alcohol", y= "speeding", data= car_crashes);
```

```python
# Scatterplot - Bivariate Distribution
sns.jointplot(x ="speeding", y= "ins_losses", data= car_crashes);
```

```python
# Scatterplot - Bivariate Distribution
sns.jointplot(x ="ins_losses", y= "speeding", data= car_crashes);
```

# Jointplot Hex

 Useful graph when working with large datasets

```python
with sns.axes_style("white"):
    sns.jointplot(x = "ins_losses",
y = "speeding",
data = car_crashes,
kind = "hex",
color = "green");
```

```python
with sns.axes_style("white"):
    sns.jointplot(x = "alcohol",
y = "speeding",
data = car_crashes,
kind = "hex",
color = "blue");
```

# Density Jointplot

```python
#Bivariate Distribution 
sns.jointplot(x= "ins_losses", y= "speeding", data= car_crashes, kind="kde")
```

```python
#Bivariate Distribution 
g = sns.jointplot(x = "ins_losses",
y = "speeding",
data = car_crashes,
kind = "kde",
color = "red")
g.plot_joint(plt.scatter, c = "w", s = 30,
linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0);
```
